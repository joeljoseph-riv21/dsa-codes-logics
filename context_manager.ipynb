*Writing a code to practice **LIFO resource principle** using **context manager** that tracks how many lines are written to a file*

```python
from contextlib import contextmanager

@contextmanager
def open_trace(filename):
    file = open(filename, "w")
    line_counter = [0]

    def write_into(sentence):
        file.write(sentence+"\n")
        line_counter[0] = line_counter[0] +  1

    try:
        yield write_into
        print(f"{line_counter[0]} lines written on {filename}")
    finally:
        file.close()
```

*using yield here a **generator function** helps to manage data only when needed*

```python
with open_trace('my_file.txt') as write:
    write("this is a sample file created on sample_data folder")
    write("joeljoseph-riv21 on github")
    write("successfully practiced context manager")
    write("life is between being lazy and letting life to be nothing easy")
```
