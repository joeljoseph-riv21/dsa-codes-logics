*Writing a program to check if the **firstletter** of a string is **unique** and not a repeating word*

```python
#separating the string
string = "finalyearstudent"
alphabets = list(string)
print(alphabets)
```

```python
#adding a memory block through a empty dictionary
compare={}
for letter in alphabets:
    if letter in compare:
        compare[letter] = compare[letter] + 1
    else:
        compare[letter] = 1
print(compare)
```

```python
#iterating through the list& a break statement gets exactly the first letter
for letter in alphabets:
    if compare[letter] > 1:
        print(f"{letter} is repeated")
    else:
        print(f"{letter} is unique")
        break
```
