*Writing code to practice number logic series*

```python
def prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        print(i)

        if num % i == 0:
           print(f"{num}%{i}")
           return False
    return True
```

```python
prime(73)
```

```python
def fibonacci(num):
    a, b = 0, 1

    for i in range(num):
        a, b = b, a + b
        print(a, end="")

        if i < num - 1:
          print(",", end="")
"""
  to print from 0 instead of 1, use print statemnt
   before a, b = b, a + b
"""
```

```python
fibonacci(9)
```

*An armstrong number is calculated based on a number's **sum of its digits, each raised to the power of the number of digits, equals the number** itself.*

```python
def armstrong(num):
    digit = str(num)
    power = len(digit)
    total_sum = 0

    for i in digit:
        total_sum = total_sum + int(i)**power

    if total_sum == num:
        print(f"{num}, an armstrong number")
    else:
        print(f"{num}, not an armstrong.")
```

```python
armstrong(777)
```
