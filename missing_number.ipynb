*Writing a code using **gauss formula** to find the missing number in an array*

```python
def get_missing(inputs):
    num = len(inputs)
    expected_sum = num * (num + 1) // 2
    actual_sum = sum(inputs)
    return expected_sum - actual_sum
```

```python
# to identify the missing value in a array, we use gauss formula finding expected sum.
# actual sum gives the exact one, expected sum begins with 1 as of length to do this.

inputs = [0, 1, 2, 3, 4, 5, 7]
print(get_missing(inputs))
```
