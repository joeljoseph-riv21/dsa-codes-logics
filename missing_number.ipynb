*Writing a code using **the grouping& categorization pattern** to practice analyst logic by filtering accounts*

```python
def filter_accounts(inputs):
    if not inputs:
       return 0.0

    notebook = []
    inactive_count = 0
    total_accounts = len(inputs)


    for account in inputs:
        if account["status"].strip().lower() == "inactive":
          notebook.append(account)
          inactive_count += 1

    rate = inactive_count / total_accounts * 100
    print (f"filtered accounts list:{notebook}\n- inactive accounts: {inactive_count}\n- churn rate: {rate}%")
```

```python
accounts = [
        {"customer_id": 101, "status": "active"},
        {"customer_id": 102, "status": "inactive"},
        {"customer_id": 103, "status": "active"},
        {"customer_id": 104, "status": "inactive"},
        {"customer_id": 105, "status": "paused"}
    ]

print(filter_accounts(accounts))
```
