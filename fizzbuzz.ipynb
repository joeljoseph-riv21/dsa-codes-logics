*Writing code for printing the multiples of 3& 5 **using fizbuzz technique***


```python
def fizbuzz_generator():
   for number in range(1, 101):
       if number % 3 == 0 and number % 5 == 0:
          yield "fizzbuzz"
       elif number % 3 == 0:
          yield "fizz"
       elif number % 5 == 0:
          yield "buzz"
       else:
          yield number
```

```python
obj = fizbuzz_generator()
print(next(obj))
print(next(obj))
print(next(obj))
print(next(obj))
print(next(obj))
print(next(obj))
print(next(obj))
print(next(obj))
print(next(obj))
```
