*Writing code to move all zeros to the end of an array using the **reader-writer** pointer strategy*

```python
def move_zero(input):
    write = 0

    for read in range(len(input)):
        if input[read] != 0:
           input[write], input[read] = input[read], input[write]
           write = write + 1

    return input
```

```python
inputs = [1, 2, 0, 3, 5, 0, 9, 7]
print(move_zero(inputs))
```
