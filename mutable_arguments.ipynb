*Writing a code to experiment **mutable default arguments** and to solve shared-reference problem*

```python
def add_to_profile(skill, profile={"skills": []}):
    profile["skills"].append(skill)
    return profile

candidate1 = add_to_profile("python")
candidate2 = add_to_profile("java")

print("candidate 1:", candidate1)
print("candidate 2:", candidate2)
```

*Inorder to handle **shared-reference problem** is to use **None** as a default value and get its own unique mutable object*

```python
def make_it_possible(skillset, profile=None):
    if profile is None:
        profile = {"skills": []}
    profile["skills"].append(skillset)
    return profile
```

```python
result = make_it_possible("pytorch")
result2 = make_it_possible("pandas")
print(f"candidate1:{result}")
print(f"candidate2:{result2}")
```
