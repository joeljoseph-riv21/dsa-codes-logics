*Writing a program to remove the punctuation and count the words occurance in a dictionary*

```python
#convert the sentence to lowercase
sentence = str(input("Enter a sentence:"))
lower = sentence.lower()
print(lower)
```

```python
#removing the punctuation from the sentence
clean_sentence = lower
punctuation = ["~","!","*","&",",",".","-","?","/"]
for i in punctuation:
    clean_sentence = clean_sentence.replace(i,"")
print(clean_sentence)
```

```python
#splitting the sentence into words as list
words = clean_sentence.split()
print(words)
```

```python
#checking frequenices of the words and storing them in a fresh dictionary
count_words = {}
for word in words:
    if word in count_words:
        count_words[word] += 1
    else:
        count_words[word] = 1
print(count_words)
```

- if word in count_words:
Inside the loop, it checks if the current word is
already a key in the count_words dictionary.
If it is, it means the word has been encountered before.
- count_words[word] += 1:
If the word exists, its corresponding value (count) is incremented by 1.
- else: count_words[word] = 1:
If the word is not yet in the dictionary, it's the first time it's been encountered, so it's added as a new key with an initial value of 1.

```python
#the dictionary contents
print("\n Final word counts \n")
for key, value in count_words.items():
    print(f"'{key}': {value} times")
```
