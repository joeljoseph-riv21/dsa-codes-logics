*Writing code using **two pointer** pattern to check if a given palindrome is valid*


```python
def check_palindrome(input):
    left_index = 0
    right_index = len(input) - 1

    while left_index < right_index:
          while left_index < right_index and not input[left_index].isalnum():
                left_index = left_index + 1

# the inner loop runs until the condiion where 'input[left_index].isalnum():'
# gets False, and goes down the second inner loop right repeat the same.
# (the logic of inner while theva iladhadha ethu, theva iladhadha erakku)

          while left_index < right_index and not input[right_index].isalnum():
                right_index = right_index - 1

# main logic
          if input[left_index].lower() == input[right_index].lower():
             left_index += 1
             right_index -= 1
          else:
             return False

    return f"[{input}] is a valid palindrome!"
```
