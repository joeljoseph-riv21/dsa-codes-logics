*Writing code with **two sum** pattern to find the indices of the two numbers that add up exactly to that target*

```python
def two_sum(input, target):
    seen = {}

    for index, value in enumerate(input):
        number = target - value
        if number in seen:
            return [seen[number], index]
        else:
            seen[value] = index
    return []

# here, seen[number] gives the index of the first number we encountered,
# and index is the index of the current number.
```

```python
inputs = [2, 21, 13, 42, 8]
targetting = 21
indecies = two_sum(inputs, targetting)
print(f"got {targetting} by adding {indecies} indecies")
```
