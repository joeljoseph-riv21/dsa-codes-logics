*Writig code to practice **multilevel inheritance, method overriding** with OOP style*

```python
class HeroStrike:
      def __init__(self, name, power, ult):
          self.name = name
          self.power = power
          self.ult = ult

      def hide(self):
          print(f"the hero {self.name} is hiding in bush")

      def strike(self):
          print(f"the hero {self.name} is to take the battery using {self.power}")

      def __str__(self) -> str:
          return f"the hero {self.name} is to take the battery using {self.power} and win."

# multi-level inheritance
class Defense(HeroStrike):
      def __init__(self, name, power, ult, seconds):
          super().__init__(name, power, ult)
          self.time = seconds

      def defend(self):
          print(f"there is {self.time} seconds left for robot charging, hero {self.name} on action.")

class Robot(Defense):
      def __init__(self, robot_name):
          self.robot_name = robot_name

# method overriding
      def defend(self):
          print(f"the robot {self.robot_name} is on action defend your tower!")
```

```python
start = HeroStrike("eldragon", "rocket plunger", "wrestle")
mid = Defense(name="mortal", power="Shield", ult="Defense Wall", seconds=7)
end = Robot("Gigatrone")
print(start)
mid.defend()
end.defend()
```
