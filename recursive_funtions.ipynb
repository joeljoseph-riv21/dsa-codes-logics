*Writing code using **recursive functions** to print the even numbers in reverse order*

```python
def reverse_printing(number, collection = None):
    if collection is None:
        collection = []

    if number < 2:
       return collection

    if number % 2 == 0:
        collection.append(number)
        return reverse_printing(number - 1, collection)
    else:
        return reverse_printing(number - 1, collection)
```

```python
reverse_printing(99)
```
