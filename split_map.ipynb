*Writing a code in **string split and conditional mapping** pattern to map files to sub-folder based on their extension*

```python
import os
import shutil

def organizer(given_path):
    for filenames in os.listdir(given_path):

        #joining the full path to check if a file or directory first
        filepath = os.path.join(given_path, filenames)
        print(f"filename:{filenames}")
        print(f"filepath:{filepath}")

        #the logic is if file proceed is directory ignore
        if os.path.isfile(filepath):
           print("[file ☑]")
        elif os.path.isdir(filepath):
           print("[directory ☑]")
           continue
        else:
           print("NULL")

        #spliiting and filtering based extensions
        _, extension = os.path.splitext(filenames)
        folder = None

        if extension in ['.txt', '.md']:
           folder = os.path.join(given_path, 'text files')
           print(f"target folder for {filenames}is{folder}")

        if extension in ['.csv']:
           folder = os.path.join(given_path, 'csv files')

        if extension in ['.json']:
           folder = os.path.join(given_path, 'json files')

        #the moving logic to respective sub-folders
        if folder:
           os.makedirs(folder, exist_ok=True)
           shutil.move(filepath, os.path.join(folder, filenames))
           print(f" moved {filenames} to {folder}")
```

```python
new = '/content/sample_data/'
organizer(new)
```
