*Writing code using logic **check-in/out counter with set()** to find the single unique ID of duplicates from the account logins in a list*

```python
def unique_id(inputs):
    tracker = set()

    for id in inputs:
        if id in tracker:
              tracker.remove(id)

        else:
              tracker.add(id)
    return f"unique account id: {(list(tracker)[0])}"
```

```python
login_ids = [31, 4, 15, 15, 4, 21, 31]
print(unique_id(login_ids))
```
