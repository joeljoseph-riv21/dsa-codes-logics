*Writing code using **variable-size expandable window** to find the length of the longest substring without repeating characters*

```python
def substring_length(inputs: str) -> int:
    track_char = set()
    left = 0
    max_length = 0

    for right in range(len(inputs)):
        while inputs[right] in track_char:
              track_char.remove(inputs[left])
              left = left + 1

        track_char.add(inputs[right])

        window_size = right - left + 1
        max_length = max(max_length, window_size)
    return max_length
```

```python
substring = "xxyyxxxbzz"
print(substring_length(substring))
```
