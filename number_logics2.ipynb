*Writing code with **iterative digit manipulation** pattern to reverse a number and summing the digits in a number*

```python
"""
inorder to reverse a number
1. extract the last digit of number by modulos
2. multiply the initialized var by 10, add the last digit
3. strictly divide the number by 10
[formula]: Modulo by 10, Multiply by 10& add, Divide by 10
"""

def reversing(num):
    reverse = 0

    while num > 0:
          last_digit = num % 10
          reverse = reverse*10 + last_digit
          num = num//10
    return reverse
```

```python
reversing(1233)
```

```python
def sum_digits(num):
    sum = 0

    while num > 0:
          last_digit = num%10
          sum = sum + last_digit
          num = num//10
    return sum
```

```python
sum_digits(4456)
```
