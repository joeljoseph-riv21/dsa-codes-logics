*Writing code for practicing **context managers** to handle file operations*


```python
from contextlib import contextmanager

@contextmanager
def file_open(filename):
    file = open(filename, "w")
    lines = [0]

    def write_file(text):
        file.write(text + "\n")
        lines[0] += 1

    try:
        yield write_file
    finally:
        file.close()
        print(f"file closed sucessfully,\ntotal lines written: {lines[0]}")
```

- The file_open function then yields the write_file inner function.
- This write_file function is then assigned to the variable write (because of as write:)  inside your with block contains a reference to the write_file function that you defined within file_open.
- Control now enters the with block.
- write is now effectively your write_file function, when you call write("joeljoseph-riv21 @github") (or any of the other calls), you are directly invoking the write_file function. The string `
