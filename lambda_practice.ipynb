*Writing code using **lambda functions** to practice logical questions*

```python
def lambdas(input_list):
    sorted_tasks = sorted(input_list, key=lambda task_item: task_item[1])
    print("task execution order:")

    for task_name, priority_value in sorted_tasks:
        print(task_name)
```

```python

task = [
    ("Loan", 3),
    ("Account Opening", 1),
    ("Transaction Approval", 2)
]

lambdas(task)
```
