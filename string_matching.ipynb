```python
paragraph = """
    Contact our team:
    Arjun:arjun@zoho.com   | 9876543210
    Priya:priya@tcs.in     | 8123456789
    Kumar:kumar@gmail.com  | 7999888777
    to test youe pattern: ram12389@567gmail
"""
```

```python
import re
def regex(para):
    pattern = r'\b[6-9]\d{9}\b'
    extracted = re.findall(pattern, para)
    print(f"phone nos.:{extracted}")

    masked = []
    for numbers in extracted:
        masking = re.sub(r'(\d{3})\d{4}(\d{3})', r'\1****\2', numbers)
        masked.append(masking)
    return f"masked nos.:{masked}"
```

```python
print(regex(paragraph))
```

```python
import re
def dates(para):
    pattern = r'(\d{2})/(\d{2})/(\d{2})'
    extracted = re.findall(pattern, para)
    print(f"extracted:{extracted}")

    #the result will be in list of tuple, converting it into date format
    formatted_dates = [f'{d[0]}/{d[1]}/{d[2]}' for d in extracted]
    return f"formatted:{formatted_dates}"
```

```python
sentence = "hey this is kanye signing in on 25/05/26, as a ML engineer and within 28/05/29 as Ops engineer"
print(dates(sentence))
```

```python
import re
def summing(para):
    new_amount = re.sub(r'\s+','', para)
    print(f"alligned amount: {new_amount}")

    extract = re.findall(r'\d+', new_amount)
    print(f"extracted amount: {extract}")

    int_converted = list(map(int, extract))
    add = sum(int_converted)
    print(f"summed amount: {add}")
```

```python
amount = "8000    rs,   9000  rs, 10000    rs"
summing(amount)
```
