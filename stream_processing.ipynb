*Writing a code to extract the first letter in a list and map to its word in a dictionary using via **frequency map and stream processing** pattern by using **list comprehension***

```python
def stream_process(lists):
    filtered = [word.lower() for word in lists if len(word)<=3]
    print(f"filtered words:{filtered}")

    my_dictionary={}
    for word in filtered:
        first_letter = word[0]
        print(f"first letter:{first_letter}")

        if first_letter not in my_dictionary:
           my_dictionary[first_letter]=[]
        my_dictionary[first_letter].append(word)
    return f"output:{my_dictionary}"
```

```python
animals=["caterpillar", "cat", "rat", "rattle snake", "fly"]
stream_process(animals)
```
