*Writing code using **one-pass greedy approach** pattern to get the best time to bug and sell a stock yielding maximum profit*

```python
def yield_profit(inputs):
    if len(inputs) < 2:
       return 0

    buying_price = float('inf')
    profit = 0

    for price in inputs:
        if price < buying_price:
           buying_price = price

        elif price - buying_price > profit:
             profit = price - buying_price

    return f"the maximum profit yielded: ₹{profit}"
```

```python
prices = [8, 9, 4, 5, 2, 7, 21]
print(yield_profit(prices))
```
