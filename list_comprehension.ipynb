*Writing code using **list comprehension** return a new list containing the squares of only the odd numbers from the original list*

```python
def list_comprehension(numbers):
    return [num ** 2 for num in numbers if num % 2 == 0]


num1 = [17, 21, 13, 4, 45, 6]
result1 = list_comprehension(num1)
print(f"{num1} results: {result1}")

num2 = [10, 15, 20, 25]
result2 = list_comprehension(num2)
print(f"{num2} results: {result2}")
```
