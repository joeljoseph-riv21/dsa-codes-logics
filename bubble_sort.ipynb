*Writing code using **brute force** pattern to effectively sort an array using bubble sort*

```python
def bubble_sort(inputs):
    size = len(inputs)

    for a in range(size - 1):
        for b in range(size - a - 1):

            if inputs[b] > inputs[b + 1]:
               inputs[b], inputs[b + 1] = inputs[b + 1], inputs[b]
    return f"sorted array: {inputs}"
```

```python
arr = [67, 34, 25, 12, 21, 13, 90, 88, 100]
print(bubble_sort(arr))
```

#### **Quick prep**

Why size - a - 1? This is the clever bit that makes the sort a little more efficient:
size - 1: This is the maximum number of comparisons we'd need in the first pass (when a=0) to reach the end of the unsorted part.
- a: In each outer loop pass (a), one element gets correctly placed at the end of the array (it's 'bubbled up'). So, for the next pass, we don't need to compare up to that already-sorted element. We can shorten our search. The - a effectively shortens the inner loop's reach by one position for every pass a completed. This avoids unnecessary comparisons with elements that are already known to be in their final sorted positions.
The additional - 1 is because we are comparing inputs[b] with inputs[b + 1]. To avoid going out of bounds (IndexError), the loop must stop one position before the very last element it could compare b against b+1 in the unsorted portion.
