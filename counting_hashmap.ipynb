*Writing code to practice **counting hashmap** pattern in a o(n) efficient manner*

```python
def check_uniqueness(words):
    unique = {}

    for char in words:
        unique[char] = unique.get(char, 0) + 1
    print(f"frequency mapping:{unique}")

    for char in words:
        if unique[char] == 1:
           print(f"unique letter:{char}")
```

```python
wording = "anddrearsean"
print(check_uniqueness(wording))
```
