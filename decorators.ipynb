```python
def my_decorator(func):

    def wrapper(*args, **kwargs):
        print("before")
        result = func(*args, **kwargs)
        print("after")
        return result
    return wrapper

def base():
    print("practicing decorators")

practice = my_decorator(base)
```

```python
def greet(name, greeting='hello'):
    print(f"{greeting}, {name}!")

greet = my_decorator(greet)
greet('Joel Joseph R')
```
