*Writing a code for **handling errors** to combine class structure, **encapsulation, and user-defined exception handling***

```python
#custom exception classes
class insufficientBalance(Exception):
  def __init__(self, balance: float, cash: float):
    self.message = f"<ERROR! YOU GOT {balance},HOW CAN YOU WITHDRAW {cash}>?"
    super().__init__(self.message)

class valueError(Exception):
  def __init__(self, amount: float):
    self.message1 = f"<ERROR! AMOUNT {amount} CANNOT BE ZERO OR NEGATIVE>"
    super().__init__(self.message1)

```

```python
#the parent class for main logic
class Bank:
  def __init__(self, name: str, initial_balance: float):
    self.name = name
    self.__balance = initial_balance

  def deposit(self, amount: float):
    if amount > 0:
      self.__balance += amount
    elif amount <= 0:
      raise valueError(amount)
    else:
      raise insufficientBalance("Deposit amount should be positive.")

  def withdraw(self, cash: float):
    if cash > 0 and cash <= self.__balance:
      self.__balance -= cash
    elif cash <= 0:
      raise valueError(cash)
    else:
      raise insufficientBalance(self.__balance, cash)

  def __str__(self):
    return f"[Customer name]: {self.name} \n[Current balance]: {self.__balance}"
```

```python
#evaluating the code
def run_atp():
    try:
      enter = Bank("Joel Joseph R", 999)
      print(enter)
      enter.withdraw(-600)
      pass
    except (insufficientBalance, valueError) as error:
      print(f"[Warning]:{error}")
    finally:
      print("[Session ended ©]")
run_atp()
```
