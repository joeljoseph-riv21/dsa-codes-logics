*Writing code using **sliding window** pattern to finding the maximum sum of a contiguous subarray  of a fixed length in a given array*

```python
# formula: new sum = previous sum - outgoing(left) + incoming(right)
def sliding_window(inputs, k):
    current_sum = sum(inputs[:k])
    max_sum = current_sum

    for i in range(len(inputs) - k):
        current_sum = current_sum - inputs[i] + inputs[i + k]

        if current_sum > max_sum:
            max_sum = current_sum

    return f"the highest sum seen: {max_sum}\nthe maximum average: {max_sum/k}"
```

```python
data = [1, 21, -5, -8, 50, 33, -2]
window = 3

print(sliding_window(data, window))
```
