```python
# printing pyramid pattern in centered
def pyramid(n):
    for number in range(1, n+1):
        space = n - number
        star = 2 * number - 1
        print(" " * space + "*" * star)

pyramid(7)
```

```python
# triangle
def triangle(n):
    for number in range(1, n+1):
        print("*" * number)

triangle(7)
```

```python
# diamond pattern: just combine pyramid's logic
def diamond(n):
    for number in range(1, n+1):
        space = n - number
        star = 2 * number - 1
        print(" " * space + "*" * star)

    for number in range(n-1, 0, -1):
        space = n - number
        star = 2 * number - 1
        print(" " * space + "*" * star)

diamond(7)
```

```python
def floyd_traingle(n):
    num = 1
    for i in range(1, n + 1):
        for j in range(i):
            print(num, end=" ")
            num = num + 1
        print()

floyd_traingle(7)
```
