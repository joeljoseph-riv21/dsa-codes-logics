*Writing code with **iterators and generators** aims to find all unique pairs of numbers from a given list and whose sum is greater than threshold, then counts it*

```python
import itertools
def generator(ids):
    for pair in itertools.combinations(ids, 2):
        yield pair

def filter(teams, threshold):
    for team in teams:
        if sum(team) > threshold:
           yield team

def display(final_teams, id):
    count=0
    for team in final_teams:
        team_sum = sum(team)
        count = count+1
        print(f"\ntotal qualifying teams:{count}, sum of team:{team_sum}")
```

```python
id = [21, 23, 25, 17, 3, 13, 5]

ids = generator(id)
filtered_ids = filter(ids, 42)
display(filtered_ids, ids)
```

```python
import itertools
permutate = list(itertools.permutations([3,2,4], 2))
print(f"permutation values:{permutate}")

slicer = list(itertools.islice(permutate, 4))
print(f"sliced values:{slicer}")

group = list(itertools.combinations(slicer, 2))
print("\tGROUPED")
for i in group:
    print(f"{str(i).center(23)}")
```
