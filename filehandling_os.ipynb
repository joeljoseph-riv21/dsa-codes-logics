*Writing a code to list all .csv files in a given folder and all its subfolders.*

```python
import os

def list_files(folder):

    for root, sub, files in os.walk(folder):
        print(f"list of files:{files}")
        print(f"list of sub-folders:{sub}")

        for extension in files:
            if extension.endswith(".csv"):
               print(f":{os.path.join(root, extension)}")
```

```python
folder_path = r"/content/sample_data"
list_files(folder_path)
```

*Writing a code to check a folder and delete any file that is empty*

```python
import os

def delete_unwanted(folder):
    for files in os.listdir(folder):
        file_path = os.path.join(folder, files)

        if os.path.isfile(file_path) and os.path.getsize(file_path) == 0:
           os.remove(file_path)
           print(f"deleted empty file: {files}")
        else:
           print(f"{files},this file is with 0 bytes")
```

```python
folder_path = r"/content/sample_data"
delete_unwanted(folder_path)
```
